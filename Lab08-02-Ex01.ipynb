## คำถาม



1. Class ในภาษา C# คืออะไร


*คำตอบ*

Class (คลาส) ในภาษา C# เป็นโครงสร้างที่ใช้ในการกำหนดรูปแบบของวัตถุ (Object) โดยกำหนดคุณสมบัติ (Properties) และพฤติกรรม (Methods) ของวัตถุนั้น

2. Class มีหน้าที่อะไร


*คำตอบ*

Class ทำหน้าที่เป็นแม่แบบ (Template) สำหรับสร้าง Object โดยใช้ในการรวบรวมข้อมูลและฟังก์ชันที่เกี่ยวข้องกันไว้ในที่เดียวกัน

3. ทำไมเราต้องใช้ Class


*คำตอบ*

Class ช่วยให้การพัฒนาโปรแกรมเป็นระเบียบและสามารถใช้หลัก OOP (Object-Oriented Programming) เช่น การห่อหุ้ม (Encapsulation), การสืบทอด (Inheritance) และการโพลีมอร์ฟิซึม (Polymorphism) ซึ่งช่วยให้โค้ดมีความยืดหยุ่นและนำกลับมาใช้ใหม่ได้ง่ายขึ้น

4. ส่วนประกอบของ Class มีอะไรบ้าง


*คำตอบ*

Fields – ตัวแปรที่ใช้เก็บข้อมูลของคลาส
Properties – ใช้สำหรับกำหนดและเข้าถึงค่าของ Fields
Methods – ฟังก์ชันที่ใช้ดำเนินการกับข้อมูลภายในคลาส
Constructors – เมธอดพิเศษที่ใช้สำหรับสร้าง Object
Events & Delegates – ใช้สำหรับการทำงานแบบ Event-driven
Modifiers – ตัวกำหนดการเข้าถึง เช่น public, private, protected

5. Class และ Object แตกต่างกันอย่างไร


*คำตอบ*

Class เป็นแม่แบบสำหรับสร้าง Object
Object เป็นอินสแตนซ์ (Instance) หรือค่าจริงที่ถูกสร้างขึ้นจาก Class

*คำตอบ*
