## คำถาม

1. Field ในภาษา C# คืออะไร


*คำตอบ*

Field คือ ตัวแปรที่ถูกประกาศภายใน class หรือ struct และใช้เก็บข้อมูลของ object

2. Field มีหน้าที่อะไรใน class 


*คำตอบ*

Field ทำหน้าที่เก็บสถานะของ object และเป็นข้อมูลที่ object ใช้งานได้

3. การประกาศ Field ทำได้อย่างไร ให้อธิบายรูปแบบการประกาศ field ในคลาส


*คำตอบ*

รูปแบบการประกาศ Field

class Person

{
    
      public string name;  

      private int age;     

      public Person(string name, int age)
      {
          this.name = name;
          this.age = age;
      }
}


4. Field และ Variable แตกต่างกันอย่างไร


*คำตอบ*

Field เป็นตัวแปรของ class ใช้เก็บข้อมูลของ object และสามารถเข้าถึงจาก method ภายใน class

Variable เป็นตัวแปรที่อยู่ภายใน method หรือ block ของ code ใช้เก็บค่าชั่วคราว

5. เมื่อไหร่ควรใช้ Field และเมื่อไหร่ควรใช้ Variable

*คำตอบ*

🔹 ควรใช้ Field เมื่อข้อมูลต้องถูกเก็บในระดับ object

🔹 ควรใช้ Variable เมื่อข้อมูลใช้เพียงชั่วคราวภายใน method
