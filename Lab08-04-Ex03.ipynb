## คำถาม

1. Access modifier ในภาษา C# คืออะไร


*คำตอบ*

Access Modifier คือ ตัวกำหนดระดับการเข้าถึง (Visibility) ของ class, field, method, property และองค์ประกอบอื่น ๆ ภายในโปรแกรม C# ซึ่งช่วยควบคุมว่าโค้ดส่วนใดสามารถเข้าถึงหรือใช้งานสมาชิกของ class ได้บ้าง

2. จงอธิบายขอบเขตและการใช้งานของ Access modifier ต่อไปนี้ 


2.1  public


*คำตอบ*

- สามารถเข้าถึงได้จาก ทุกที่ ทั้งภายในและภายนอก class หรือ assembly

- ใช้งานได้จาก class อื่น ๆ โดยไม่มีข้อจำกัด

2.2 private


*คำตอบ*

- สามารถเข้าถึงได้ เฉพาะภายใน class นั้น ๆ เท่านั้น

- class อื่น ๆ ไม่สามารถเข้าถึงได้ แม้จะอยู่ในไฟล์เดียวกัน

2.3 protected


*คำตอบ*

- สามารถเข้าถึงได้เฉพาะ class นั้น และ class ที่สืบทอด (inherit) จากมัน

- ไม่สามารถเข้าถึงจาก class อื่น ๆ ที่ไม่ใช่ลูกหลานของ class นั้น

2.4 internal


*คำตอบ*

- สามารถเข้าถึงได้เฉพาะ ภายใน assembly เดียวกัน

- ไม่สามารถเข้าถึงจาก assembly อื่น ๆ

2.5 protected internal

*คำตอบ*

- สามารถเข้าถึงได้ ภายใน assembly เดียวกัน

- สามารถเข้าถึงได้จาก class ลูก ที่สืบทอดจาก class นั้น แม้จะอยู่คนละ assembly**

3. เครื่องหมาย `?` ในบรรทัดที่ 13 `protected string? name;` มีความหมายว่าอย่างไร  

*คำตอบ*

? หลัง string หมายความว่า ตัวแปร name สามารถมีค่า null ได้ หรือเรียกว่า Nullable Reference Type ในภาษา C#
