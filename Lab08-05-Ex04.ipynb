## คำถาม

1. Field ในภาษา C# คืออะไร


*คำตอบ*

Field คือ ตัวแปรที่ถูกประกาศภายใน class หรือ struct และใช้เก็บข้อมูลของ object

2. Field มีหน้าที่อะไรใน class 


*คำตอบ*

Field ทำหน้าที่เก็บสถานะของ object และเป็นข้อมูลที่ object ใช้งานได้

3. การประกาศ Field ทำได้อย่างไร

*คำตอบ*

รูปแบบการประกาศ Field

class Person

{
    
      public string name;  

      private int age;     

      public Person(string name, int age)
      {
          this.name = name;
          this.age = age;
      }
}


4. Implicit fields initialization  คืออะไร

*คำตอบ*

Implicit Fields Initialization คือการกำหนดค่าเริ่มต้นให้กับฟิลด์ (fields) ของอ็อบเจ็กต์โดยอัตโนมัติ โดยไม่ต้องกำหนดค่าเริ่มต้นให้กับแต่ละฟิลด์อย่างชัดเจนในโค้ด

5. Explicit field initialization คืออะไร



*คำตอบ*

Explicit Field Initialization คือการกำหนดค่าเริ่มต้นให้กับฟิลด์ (fields) อย่างชัดเจนภายในคลาสหรือโครงสร้างข้อมูล (struct) โดยกำหนดค่าให้ตัวแปรตั้งแต่ตอนประกาศ หรือใน constructor ของคลาส

6. Implicit fields initialization  และ Explicit field initialization แตกต่างกันอย่างไร



*คำตอบ*

Implicit Field Initialization และ Explicit Field Initialization แตกต่างกันตรงที่วิธีการกำหนดค่าเริ่มต้นให้กับตัวแปรหรือฟิลด์ในคลาส

Implicit Field Initialization คือการที่ค่าของตัวแปรถูกกำหนดโดยอัตโนมัติตามค่าปริยายของภาษาโปรแกรม ตัวอย่างเช่น ใน Java ตัวแปรประเภทตัวเลขจะได้รับค่า 0 และตัวแปรประเภท boolean จะได้รับค่า false โดยไม่ต้องกำหนดค่าเอง แต่ใน C++ ตัวแปรอาจไม่ได้รับค่าปริยายและอาจมีค่าขยะหากไม่ได้กำหนดค่า

Explicit Field Initialization คือการที่โปรแกรมเมอร์กำหนดค่าเริ่มต้นให้กับตัวแปรโดยตรง เช่น int x = 10; หรือการใช้ constructor เพื่อกำหนดค่าให้กับตัวแปรเมื่อสร้างอ็อบเจ็กต์ วิธีนี้ช่วยให้มั่นใจได้ว่าค่าของตัวแปรจะเป็นไปตามที่ต้องการและลดโอกาสเกิดข้อผิดพลาดจากค่าขยะ
