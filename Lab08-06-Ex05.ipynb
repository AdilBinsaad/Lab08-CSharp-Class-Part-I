*คำตอบ*

- F1 (int) และ F2 (string) ไม่ได้กำหนดค่าเริ่มต้น → C# กำหนดค่า implicit initialization (F1 = 0, F2 = null) และแจ้งเตือนว่า F2 อาจเป็น null

- F3 และ F4 ถูกกำหนดค่าโดยตรง → เป็น explicit initialization (F3 = 100, F4 = "ASDF")

- คำเตือน CS8618 และ CS0649 บอกว่า F2 อาจเป็น null และ F1 ไม่ได้รับค่าก่อนใช้งาน

โค้ดรันได้สำเร็จ แต่อาจต้องใช้ required หรือกำหนดค่าให้ F1, F2 ใน constructor เพื่อลดคำเตือน

8. นำ code ในข้อ 2 มารันบน polyglot notebook  โดยใส่ใน  code cell ด้านล่างนี้


````
var myObj = new MyClass();
 myObj.Info();
 
 class MyClass
 {
    int F1;
    string F2;
    int F3 = 100;
    string F4 = "ASDF";

     public void Info()
     {
         System.Console.WriteLine($"Default integer      F1 = {F1}");  //Implicit fields initialization
         System.Console.WriteLine($"Default string       F2 = {F2}");
 
         System.Console.WriteLine($"Initialized integer  F3 = {F3}");  //Explicit field initialization
         System.Console.WriteLine($"Initialized string   F4 = {F4}");
     } 
 }
````
