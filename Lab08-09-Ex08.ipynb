## คำถาม

1. static class คืออะไร และแตกต่างจาก class ธรรมดาอย่างไร

*คำตอบ*

static class คือคลาสที่ไม่สามารถสร้างอ็อบเจ็กต์ (instance) จากมันได้ และสมาชิกทั้งหมดในคลาสนี้ต้องเป็น static

ความแตกต่างจาก class ธรรมดา

- static class: ใช้เพื่อจัดเก็บเมธอดหรือฟิลด์ที่ไม่ต้องการการสร้างอ็อบเจ็กต์

- class ธรรมดา: สามารถสร้างอ็อบเจ็กต์และมีทั้ง static และ non-static สมาชิกได้

2. เมื่อไหร่ควรใช้ static class และเมื่อไหร่ควรใช้ class ธรรมดา

*คำตอบ*

ใช้เมื่อคุณต้องการกลุ่มฟังก์ชันหรือค่าคงที่ที่ไม่จำเป็นต้องมีการสร้างอ็อบเจ็กต์ เช่น utility functions หรือ helper methods

3. สมาชิกทั้งหมดของ static class ต้องเป็น static หรือไม่ เพราะอะไร

*คำตอบ*

สมาชิกทั้งหมดของ static class ต้องเป็น static: ใช่ เพราะ static class ไม่มีการสร้างอ็อบเจ็กต์ ดังนั้นทุกอย่างต้องเป็น static เพื่อเข้าถึงได้โดยตรงจากคลาส

4. เราสามารถสร้าง instance ของ static class ได้หรือไม่  เพราะอะไร

*คำตอบ*

สามารถสร้าง instance ของ static class ได้หรือไม่: ไม่สามารถทำได้ เพราะ static class ออกแบบมาเพื่อไม่ให้สร้างอ็อบเจ็กต์
